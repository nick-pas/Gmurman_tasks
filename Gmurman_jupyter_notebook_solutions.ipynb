А- сработал первый сигнализатор

$$ A\overline{B} -либо- В\overline{A}$$

партии из четырех? б) выиграть не менее двух партий\
из четырех или не менее трех партий из пяти? Ничьи\
во внимание не принимаются.

результат эксперимента подчиняются распределению Бернулли, при том что соперники равносильны 
p=1/2\

так как последовательность неважна в данных событиях, то вероятность наступления события можно расссчитать черз формулу Бернулли (биномиальное распределение)
$$P=C_n^kp^kq^{n-k}$$
